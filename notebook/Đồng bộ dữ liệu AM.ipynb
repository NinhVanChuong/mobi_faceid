## Load dữ liệu AM

```python
am_df = pd.read_excel('DS_AMGP.xlsx',)
am_df = am_df[['name','emp_code_mobifone','Tỉnh/TP']]
# Tạo dictionary mapping từ tên tỉnh sang ID
dept_mapping = {
    'HPH': '126',  # AM Hải Phòng
    'BGI': '131',  # AM Bắc Giang
    'BKA': '134',  # AM Bắc Kạn
    'BNI': '130',  # AM Bắc Ninh
    'CBA': '133',  # AM Cao Bằng
    'HYE': '129',  # AM Hưng Yên
    'HDU': '128',  # AM Hải Dương
    'LSO': '135',  # AM Lạng Sơn
    'QNI': '127',  # AM Quảng Ninh
    'TBI': '132',  # AM Thái Nguyên
    'TNG': '136'   # AM Thái Bình
}

# Thêm cột department_id vào dataframe
am_df['department_id'] = am_df['Tỉnh/TP'].map(dept_mapping)
am_df['job_title'] = 'AMGP'
am_df = am_df[['name','emp_code_mobifone','job_title','department_id']]
am_df
```

```python
am_df = am_df.to_dict('records')
create_result = odooApi.execute_kw(database,uid,password,'hr.employee','create',[am_df])
```
